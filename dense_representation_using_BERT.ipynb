```python
import spacy
from sklearn.metrics.pairwise import cosine_similarity

# Example documents
documents = [
    "The Amazon rainforest is a diverse ecosystem.",
    "Tigers are often found in dense jungles.",
    "Forests play a crucial role in maintaining ecological balance."
]

# Input sentence
input_sentence = input("Enter a sentence: ")

# Load spaCy model with word embeddings
nlp = spacy.load("en_core_web_md")

# Get the embeddings for the input sentence and each document
embedding_sentence = nlp(input_sentence).vector
document_embeddings = [nlp(doc).vector for doc in documents]

# Calculate cosine similarity between input sentence and document embeddings
similarity_scores = [cosine_similarity([embedding_sentence], [doc_embedding])[0][0] for doc_embedding in document_embeddings]

# Retrieve the document with the highest similarity
most_similar_document_index = similarity_scores.index(max(similarity_scores))
most_similar_document = documents[most_similar_document_index]

print(f"Input Sentence: '{input_sentence}'")
print(f"Most similar document: '{most_similar_document}'")
```
